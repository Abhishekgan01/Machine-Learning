```python
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y , random_state=104,test_size=0.2, shuffle=True)
```

```python
print('')
print('y_test : ')
print(y_test.info)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
print(lm.intercept_)
```

```python
lm.coef_ #It tells you how much the target variable (y) will change when a specific feature (X) changes by 1 unit, while keeping all other features constant.
```

```python
X_train.columns
```

```python
cdf = pd.DataFrame(lm.coef_,X.columns,columns=['Coeff'])
cdf
```

```python
X_test
```

```python
predictions = lm.predict(X_test)
predictions
```

```python
data = {'Actual Values': y_test, 'Predicted Values': predictions}

df = pd.DataFrame(data)

print(df)
```

```python
plt.scatter(y_test, predictions)

plt.xlabel('y_test Values')
plt.ylabel('Predicted Values')
plt.title('y_test vs Predicted Values')

plt.show()
```

```python
sns.displot(y_test-predictions)
```

```python
from sklearn import metrics
metrics.mean_absolute_error(y_test,predictions) #measures the average of the absolute differences between the actual values and the predicted values.
```

```python
metrics.mean_squared_error(y_test,predictions)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,predictions))
```
